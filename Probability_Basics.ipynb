Question 4: The population of a school is divided into 60% boys and 40% girls. If you want equal representation
of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling?
Why?

Answer: For the situation described, we should use Stratified Sampling.

Targeted Representation: Stratified sampling allows you to divide the population into specific subgroups (strata)—in this case, boys and girls. You can then deliberately select an equal number from each group to ensure the sample is 50% male and 50% female, regardless of the population's actual 60/40 split.

Reduced Sampling Error: Simple Random Sampling relies on chance. In a 60/40 population, a random draw is statistically more likely to produce a sample that reflects that 60/40 split (or something close to it) rather than the 50/50 "equal representation" you specifically desire.

Precision: By "stratifying" the population first, you guarantee that the specific characteristic you are interested in (gender) is represented exactly as you intended, which makes your data more reliable for comparing the two groups.

Question 10: Differentiate between discrete and continuous random variables with examples.

Answer: The key difference between discrete and continuous random variables lies in the nature of their possible values and how they are measured.

Discrete Random Variables

These variables have a countable number of possible outcomes. They typically represent things that are counted in whole numbers, where there are distinct "gaps" between values.

Values: Finite or countably infinite (e.g., 0,1,2...)

Examples:
The number of children in a family.

The outcome of rolling a six-sided die (1,2,3,4,5,6)

Continuous Random Variables

These variables can take on any value within a given range or interval. They represent measurements that can be infinitely precise, with no gaps between possible values.

Values: Uncountable/Infinite (can include decimals and fractions).

Examples:
The exact height of students in a classroom (e.g.,  145 cm).

The time it takes for a lightbulb to burn out.
